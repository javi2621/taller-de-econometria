2.1. **(15 puntos) Genere un modelo de regresión lineal, sobre alguna de las variables dependientes,
con respecto a las variables sales, roe y ros. Interprete sus resultados y la significancia de los

```r
modelo <- lm ( salary ~ sales + roe + ros , data = ceosal1 )
summary ( modelo )
```

### Respuesta 2.1

- **Intercepto (B0):** representa el salario esperado (en miles de dólares) cuando ventas, roe y ros son cero.

- **sales(B1):** Por cada millón de dólares adicional en ventas, se espera que el salario del CEO aumente en B1 miles de dólares, manteniendo roe y ros constantes.

- **roe (B2):** Por cada 1%  adicional en el retorno sobre el capital, el salario cambia en B2 miles de dólares.

- **ros (B3):** Por cada 1% en el retorno sobre las acciones, el salario cambia en B3 miles de dólares.

**Significancia individual:**  
- Si p < 0.05, la variable es significativa al 5%.
- Si p > 0.05, no se rechaza la hipótesis nula de que el coeficiente es cero.

**Significancia global del modelo al 5%:**  
Al final del summary se reporta el estadístico F y su p-valor.  
Si el p-valor del F es menor a 0.05, se rechaza la hipótesis nula de que todos los coeficientes son iguales a cero, excepto el intercepto.
Por lo que, el modelo es globalmente significativo.

2.3 **(15 puntos) Realiza un test de hipotesis para evaluar si las dos variables que se agregaron son significativas para el modelo simultaneamente. Use significancia del 5%.**

```r
F_critico <- qf(0.95, 2, 203)
R2_r <- 0.03191
R2_ur <- 0.04059
F_est <- (R2_ur - R2_r) / (1 - R2_ur) * (203 / 2)

if (F_est > F_critico) {
  "Se rechaza H0"
} else {
  "No se rechaza H0"
}
```

####Respuesta 2.3

- **H0:** Los coeficientes de sales_mayor_promedio y sales_mayor_promedio_roe son simultáneamente iguales a cero, por lo que no aportan al modelo.
- **H1:** Al menos uno de ellos es distinto de cero.

Con un F crítico al 5% de r F_critico y un F estimado de r F_est , si F_est > F_critico, se rechaza H0.

Esto indica que las variables agregadas en el inciso 1.2 y 1.4, en conjunto, son estadísticamente significativas para explicar el salario.

2.4 **(15 puntos) Volviendo a la regresión del salario en función de sales, roe y ros. Considere
dos grupos, Grupo 1: Aquellos clientes que sales es mayor al promedio, y Grupo 2: Aquellos
clientes que sales es menor al promedio. Realice un test de hipotesis para grupos al 5% de
significancia. Use el metodo que más le acomode.**

```r
# haremos c o m p a r a c i n de grupos mediante test F de significancia de,→ variables
# para esto debemos crear variables que estan relacionadas con la dummy

ceosal1$sales_D <- ceosal1$sales * ceosal1$sales_mayor_promedio
ceosal1$roe_D  <- ceosal1$roe  * ceosal1$sales_mayor_promedio
ceosal1$ros_D  <- ceosal1$ros  * ceosal1$sales_mayor_promedio
modelo_ur <- lm(salary ~ sales + roe + ros + sales_mayor_promedio + sales_D + roe_D + ros_D, data = ceosal1)
summary(modelo_ur)  # para ver coeficientes

F_critico <- qf(0.95, 4, 201)
R2_ur <- 0.06028
R2_r <- 0.03191
F_est <- (R2_ur - R2_r) / (1 - R2_ur) * (201 / 4)

if (F_est > F_critico) {
  "Se rechaza H0: los modelos de los dos grupos son diferentes"
} else {
  "No se rechaza H0: los modelos son iguales"
}
```

#### Respuesta 2.4

Este test evalúa si la relación entre salario y las variables sales, roe y ros es estructuralmente distinta para las empresas con ventas superiores al promedio (Grupo 1) y para las que están por debajo (Grupo 2).

- **H0:** Los coeficientes de todas las variables son iguales en ambos grupos (no hay cambio estructural).
- **H1:** Al menos un coeficiente difiere entre los dos grupos.

Usamos el test F basado en la comparación de los R² del modelo restringido (sin interacciones) y el no restringido (con dummy e interacciones).

Con 4 grados de libertad y alfa = 5%, si F_est > F_critico, se rechaza H0.

Eso significa que el efecto de las variables explicativas sobre el salario cambia dependiendo de si la empresa tiene ventas altas o bajas.

Por lo tanto, sería inapropiado usar un único modelo para toda la muestra.

2.5 **(15 puntos) Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales.**

```r
ceosal1_indus <- subset(ceosal1, indus == 1)
ceosal1_finance <- subset(ceosal1, finance == 1)
ceosal1_conprod <- subset(ceosal1, consprod == 1)
ceosal1_utility <- subset(ceosal1, utility == 1)

modelo_pool <- lm(salary ~ sales + roe + ros, data = ceosal1)
modelo_indus <- lm(salary ~ sales + roe + ros, data = ceosal1_indus)
modelo_finance <- lm(salary ~ sales + roe + ros, data = ceosal1_finance)
modelo_conprod <- lm(salary ~ sales + roe + ros, data = ceosal1_conprod)
modelo_utility <- lm(salary ~ sales + roe + ros, data = ceosal1_utility)

SSR_r <- sum(residuals(modelo_pool)^2)
SSR_indus <- sum(residuals(modelo_indus)^2)
SSR_finance <- sum(residuals(modelo_finance)^2)
SSR_conprod <- sum(residuals(modelo_conprod)^2)
SSR_utility <- sum(residuals(modelo_utility)^2)
SSR_ur <- SSR_indus + SSR_finance + SSR_conprod + SSR_utility

G <- 4
k <- 3
n <- nrow(ceosal1)
gl1 <- (G - 1) * (k + 1)
gl2 <- n - G * (k + 1)

F_critico <- qf(0.95, gl1, gl2)
F_est <- (SSR_r - SSR_ur) / SSR_ur * (gl2 / gl1)

if (F_est > F_critico) {
  "Se rechaza H0: los modelos para los cuatro sectores son diferentes"
} else {
  "No se rechaza H0: los modelos sectoriales son iguales"
}
```

####2.5 Respuesta

Se aplica el test de Chow para determinar si los coeficientes del modelo salary ~ sales + roe + ros son idénticosen los cuatro sectores: industrial, financiero, productos de consumo y servicios públicos.

- **H0:** Los parámetros son los mismos en todos los sectores (un solo modelo es suficiente).

- **H1:** Al menos un sector tiene parámetros distintos.

Se hace uso de la suma de residuos al cuadrado del modelo agrupado (SSR_r) y de los cuatro modelos individuales (SSR_ur).

Asi mismo, el estadístico F se compara con el valor crítico de una F con gl1 y gl2 grados de libertad.

Si F_est > F_critico, rechazamos H0, lo que indica que el tipo de sector influye significativamente en la relación salario-desempeño.

Por lo que, conviene estimar modelos separados para cada sector. Ya que, en caso contrario no hay evidencia suficiente para afirmar diferencias estructurales, y un modelo único sería aceptable.
